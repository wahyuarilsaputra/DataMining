````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from collections import OrderedDict
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification
from sklearn.svm import SVC
````

````python
dataset = "https://raw.githubusercontent.com/wahyuarilsaputra/dataset/main/Crop_recommendation.csv"
data = pd.read_csv(dataset)
data
````

# Kategori to Numerik

````python
dataLabel = data[["label"]]
````

````python
numerikLabel = pd.get_dummies(dataLabel)
numerikLabel
````

#Normalisasi

````python
scalerN = MinMaxScaler()
scalerP = MinMaxScaler()
scalerK = MinMaxScaler()
scalerTemperatur = MinMaxScaler()
scalerHumidity = MinMaxScaler()
scalerPh = MinMaxScaler()
scalerRainfall = MinMaxScaler()
scalerData = MinMaxScaler()
````

````python
dataNormalisasi = data.drop(columns=['label'])
dataNormalisasi
````

````python
dataN = dataNormalisasi[['N']]
dataP = dataNormalisasi[['P']]
dataK = dataNormalisasi[['K']]
dataTemperatur = dataNormalisasi[['temperature']]
dataHumidity = dataNormalisasi[['humidity']]
dataPh = dataNormalisasi[['ph']]
dataRainfall = dataNormalisasi[['rainfall']]
````

````python
#normalisasiN
FrameN = scalerN.fit_transform(dataN.values)
hasilN = scalerN.transform(FrameN)
hasilN = scalerN.transform(np.array([[90]]))
hasilN
````

````python
#normalisasiP
FrameP = scalerP.fit_transform(dataP.values)
hasilP = scalerP.transform(FrameP)
hasilP = scalerP.transform(np.array([[54]]))
hasilP
````

````python
#normalisasiK
FrameK = scalerK.fit_transform(dataK.values)
hasilK = scalerK.transform(FrameK)
hasilK = scalerK.transform(np.array([[16]]))
hasilK
````

````python
#normalisasiTemperatur
FrameTemperatur = scalerTemperatur.fit_transform(dataTemperatur.values)
hasilTemperatur = scalerTemperatur.transform(FrameTemperatur)
hasilTemperatur = scalerTemperatur.transform(np.array([[22.61359953]]))
hasilTemperatur
````

````python
#normalisasiHumidity
FrameHumidity = scalerHumidity.fit_transform(dataHumidity.values)
hasilHumidity = scalerHumidity.transform(FrameHumidity)
hasilHumidity = scalerHumidity.transform(np.array([[63.69070564]]))
hasilHumidity
````

````python
#normalisasiPh
FramePh = scalerPh.fit_transform(dataPh.values)
hasilPh = scalerPh.transform(FramePh)
hasilPh = scalerPh.transform(np.array([[5.7499144210000015]]))
hasilPh
````

````python
#normalisasiRainfall
FrameRainfall = scalerRainfall.fit_transform(dataRainfall.values)
hasilRainfall = scalerRainfall.transform(FrameRainfall)
hasilRainfall = scalerRainfall.transform(np.array([[87.75953857]]))
hasilRainfall
````

````python
#NormalisasiSeluruhData
FramehasilNormalisasi = scalerData.fit_transform(dataNormalisasi.values)
````

````python
FramehasilNormalisasi
````

````python
FramehasilNormalisasi = pd.DataFrame(FramehasilNormalisasi)
FramehasilNormalisasi.columns = ['N','P','K','temperature','humidity','ph','rainfall']
FramehasilNormalisasi
````

````python
label = data['label']
dataHasil = pd.concat([label,FramehasilNormalisasi],axis = 1)
dataHasil
````

````python
percent_amount_of_test_data = 0.3
````

````python
matrices_X = dataHasil.iloc[:,1:8].values
matrices_Y = dataHasil.iloc[:,0].values
````

````python
X1 = dataHasil.iloc[:,1:8].values
Y1 = dataHasil.iloc[:, 0].values
````

````python
X_train, X_test, y_train, y_test = train_test_split(X1, Y1, test_size = percent_amount_of_test_data, random_state=0)
````

````python
### Dictionary to store model and its accuracy

model_accuracy = OrderedDict()

### Dictionary to store model and its precision

model_precision = OrderedDict()

### Dictionary to store model and its recall

model_recall = OrderedDict()
````

````python
#Naive Bayes
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
Y_pred_nb = naive_bayes_classifier.predict(X_test)

cm = confusion_matrix(y_test, Y_pred_nb)

print('Confusion matrix for Gaussian Naive Bayes\n',cm)

naive_bayes_accuracy = round(100 * accuracy_score(y_test, Y_pred_nb), 2)
model_accuracy['Gaussian Naive Bayes'] = naive_bayes_accuracy

naive_bayes_precision = round(100 * precision_score(y_test, Y_pred_nb, average = 'weighted'), 2)
model_precision['Gaussian Naive Bayes'] = naive_bayes_precision

naive_bayes_recall = round(100 * recall_score(y_test, Y_pred_nb, average = 'weighted'), 2)
model_recall['Gaussian Naive Bayes'] = naive_bayes_recall

print('The accuracy of this model is {} %.'.format(naive_bayes_accuracy))
print('The precision of this model is {} %.'.format(naive_bayes_precision))
print('The recall of this model is {} %.'.format(naive_bayes_recall))
````

````python
#KNN
K = 10
knn = KNeighborsClassifier(n_neighbors = K)

knn.fit(X_train,y_train)
````

````python
Y_pred_knn = knn.predict(X_test)
````

````python
knn.score(X_test, y_test)
````

````python
### Making the confusion matrix
cm = confusion_matrix(y_test, Y_pred_knn)

### Printing the accuracy, precision, and recall of the model
print('Confusion matrix for K - Nearest Neighbors\n',cm)
nn1_accuracy = round(100 * accuracy_score(y_test, Y_pred_knn), 2)
model_accuracy['1 - Nearest Neighbors'] = nn1_accuracy

nn1_precision = round(100 * precision_score(y_test, Y_pred_knn, average = 'weighted'), 2)
model_precision['1 - Nearest Neighbors'] = nn1_precision

nn1_recall = round(100 * recall_score(y_test, Y_pred_knn, average = 'weighted'), 2)
model_recall['1 - Nearest Neighbors'] = nn1_recall

print('The accuracy of this model is {} %.'.format(nn1_accuracy))
print('The precision of this model is {} %.'.format(nn1_precision))
print('The recall of this model is {} %.'.format(nn1_recall))
````
