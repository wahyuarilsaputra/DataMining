```python
dataset = "https://raw.githubusercontent.com/wahyuarilsaputra/dataset/main/weatherAUS.csv"
data = pd.read_csv(dataset)
data
```

```python
dataLokasi = data[["Location"]]
dataWind1 = data[["WindGustDir"]]
dataWind2 = data[["WindDir9am"]]
dataWind3 = data[["WindDir3pm"]]
dataToday = data[["RainToday"]]
dataTomorrow = data[["RainTomorrow"]]
```

#Normalisasi

```python
scaler = MinMaxScaler()
```

```python
dataNormalisasi = data.drop(columns=['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'])
dataNormalisasi
```

```python
FramehasilNormalisasi = scaler.fit_transform(dataNormalisasi.values)
FramehasilNormalisasi
```

```python
FramehasilNormalisasi = pd.DataFrame(FramehasilNormalisasi)
FramehasilNormalisasi.columns = ['MinTemp',	'MaxTemp',	'Rainfall',	'Evaporation'	,'Sunshine','WindGustSpeed','WindSpeed9am',	'WindSpeed3pm',	'Humidity9am',	'Humidity3pm',	'Pressure9am',	'Pressure3pm',	'Cloud9am',	'Cloud3pm',	'Temp9am',	'Temp3pm']
FramehasilNormalisasi
```

```python
hapus= FramehasilNormalisasi.dropna(axis=0)
hapus
```

```python
label = data['RainToday']
label2 = data['RainTomorrow']
Hasil = pd.concat([label,label2,hapus,FramehasilNormalisasi], axis=1)
Hasil = Hasil.dropna(axis=0)
Hasil
```

```python
percent_amount_of_test_data = 0.3
```

```python
matrices_X = Hasil.iloc[:,2:34].values
matrices_Y = Hasil.iloc[:,0:1].values
```

```python
X1 = Hasil.iloc[:,2:34].values
Y1 = Hasil.iloc[:, 0:1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, Y1, test_size = percent_amount_of_test_data, random_state=0)
```

```python
### Dictionary to store model and its accuracy

model_accuracy = OrderedDict()

### Dictionary to store model and its precision

model_precision = OrderedDict()

### Dictionary to store model and its recall

model_recall = OrderedDict()
```

```python
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
Y_pred_nb = naive_bayes_classifier.predict(X_test)

cm = confusion_matrix(y_test, Y_pred_nb)

print('Confusion matrix for Gaussian Naive Bayes\n',cm)

naive_bayes_accuracy = round(100 * accuracy_score(y_test, Y_pred_nb), 2)
model_accuracy['Gaussian Naive Bayes'] = naive_bayes_accuracy

naive_bayes_precision = round(100 * precision_score(y_test, Y_pred_nb, average = 'weighted'), 2)
model_precision['Gaussian Naive Bayes'] = naive_bayes_precision

naive_bayes_recall = round(100 * recall_score(y_test, Y_pred_nb, average = 'weighted'), 2)
model_recall['Gaussian Naive Bayes'] = naive_bayes_recall

print('The accuracy of this model is {} %.'.format(naive_bayes_accuracy))
print('The precision of this model is {} %.'.format(naive_bayes_precision))
print('The recall of this model is {} %.'.format(naive_bayes_recall))
```
